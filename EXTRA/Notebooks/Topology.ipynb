```python
# Parametrisation following the source mentioned above for the shower peak position for each energy
def parametrisation(E_L):
    P = -41.50 + 113.9 * E_L - 4.103 * E_L**2
    return P
```

```python
# define your list with energies
E = np.linspace(5, 13, 100)

# initialise a list that stores the shower peak position
P = []

# assign a position to each energy
for i in range(0, len(E)):
    E_L = E[i]
    p = parametrisation(E_L)
    P.append(p)
```

```python
# Plotting
plt.figure(figsize=(10, 6))
plt.plot(E, P, color="red", linestyle="solid")

plt.title('Shower Peak Position vs Energy')
plt.xlabel('$Log_{10}$ E [GeV]')
plt.ylabel('Shower Peak Position [g $cm^{-2}$]')
plt.xlim(5, 13)
plt.ylim(200, 1000)
plt.grid(True)
plt.show()
```

```python
# look up the water densitiy for the Tanganyika site and the AMADEUS site
rho_T = 0.998                    # g cm-3
rho_A = 1.03                     # g cm-3
```

```python
# Initialize lists to store the maximum depth in meters
D_T = []
D_A = []


# assign a position in meter by dividing by the density to each energy
for i in range(0, len(E)):
    E_L = E[i]
    p = parametrisation(E_L)
    DisT = (p / rho_T) / 100
    DisA = (p / rho_A) / 100
    D_T.append(DisT)
    D_A.append(DisA)
```

```python
# Plotting
plt.figure(figsize=(10, 6))
plt.plot(E, D_T, color="blue", linestyle="solid", label="Lake Tanganyika")
plt.plot(E, D_A, color="red", linestyle="solid", label="AMADEUS Site")


plt.title('Shower Peak Position vs Energy')
plt.xlabel('$Log_{10}$ E [GeV]')
plt.ylabel('Shower Peak Position [m]')
plt.xlim(5, 13)


plt.legend()
plt.grid(True)
plt.show()
```

```python
log_energy = np.linspace(7, 17, 100)
neutrino_energy = [10**i for i in log_energy]

def relation(neutrino_energy):
    pressure = 2.12 * 10**(-11) * neutrino_energy
    pressure_mPa = pressure * 1000
    return pressure_mPa
```

```python
plot_pressure = []

for i in range(0, len(neutrino_energy)):
    energy = neutrino_energy[i]
    plot_pressure.append(relation(energy))
```

```python
energy_threshold = 5 / (1000 * 2.12 * 10**(-11))

log_energy_threshold = m.log10(energy_threshold)

# Plotting
plt.figure(figsize=(10, 6))
plt.loglog(neutrino_energy, plot_pressure, color="red", linestyle="solid")

plt.title('Pulse Pressure vs Energy')
plt.xlabel('Energy [GeV]')
plt.ylabel('$Pressure$ [mPa]')

plt.axhline(10000, color="orange", alpha=0.5, label="Want to Find")

plt.axhline(5, color="blue", alpha=0.5, label="Threshold Pressure = 5 mPa")
plt.axvline(energy_threshold, color="blue", linestyle="dashed", alpha=0.5, label=f"Minimum Measurable Energy = {energy_threshold} GeV")

plt.grid(True)
plt.legend()
plt.show()
```
