```python
import matplotlib.pyplot as plt
```

## Distance vs. Elevation

25000 Hz

Hill-up

```python
height_receiver = [0.5, 1.0, 1.5, 2.0, 2.5, 3.0]
transmittable_distance_min25u = [ 600, 850, 1000, 1200, 1300, 1450]
transmittable_distance_max25u = [ 650, 900, 1050, 1250, 1350, 1500]
```

25000 Hz

Hill-down

```python
transmittable_distance_min25d = [ 600, 850, 1000, 1150, 1300, 1450]
transmittable_distance_max25d = [ 650, 900, 1050, 1200, 1350, 1500]
```

45000 Hz

Hill-up

```python
transmittable_distance_min45u = [ 600, 850, 1000, 1200, 1300, 1450]
transmittable_distance_max45u = [ 650, 900, 1050, 1250, 1350, 1500]
```

45000 Hz

Hill-down

```python
transmittable_distance_min45d = [ 600, 850, 1000, 1150, 1300, 1450]
transmittable_distance_max45d = [ 650, 900, 1050, 1200, 1350, 1500]
```

```python
plt.fill_between(height_receiver, transmittable_distance_min25u, transmittable_distance_max25u, color='orange', alpha=0.4)
plt.title('Slope Down 25 kHz')

plt.xlabel('height of the receiver/transmitter [m]')
plt.ylabel('distance the acoustic signal can travel [m]')
```

```python
plt.fill_between(height_receiver, transmittable_distance_min25d, transmittable_distance_max25d, color='green', alpha=0.2)
plt.title('Slope Up 25 kHz')

plt.xlabel('height of the receiver/transmitter [m]')
plt.ylabel('distance the acoustic signal can travel [m]')
```
