The depth of the pancake is made proportional with the energy in the PanDepth files. The NMin value is varied between 4, 10, and 20 for comparison.

```python
log_min_energy = 18
log_max_energy = 24

# Filter out the energy values that have been detected so have flag ==1
detected_energy_P = [energy for energy, flag in zip(energy_P20, flags_P20) if flag == 1]
detected_energy_P10 = [energy for energy, flag in zip(energy_P10, flags_P10) if flag == 1]
detected_energy_P4 = [energy for energy, flag in zip(energy_P4, flags_P4) if flag == 1]
detected_energy_T = [energy for energy, flag in zip(energy_T20, flags_T20) if flag == 1]
detected_energy_T10 = [energy for energy, flag in zip(energy_T10, flags_T10) if flag == 1]
detected_energy_T4 = [energy for energy, flag in zip(energy_T4, flags_T4) if flag == 1]
detected_energy_L = [energy for energy, flag in zip(energy_L20, flags_L20) if flag == 1]
detected_energy_L10 = [energy for energy, flag in zip(energy_L10, flags_L10) if flag == 1]
detected_energy_L4 = [energy for energy, flag in zip(energy_L4, flags_L4) if flag == 1]

# Define the bin edges equally spaced in log space
log_bin_edges = np.linspace(log_min_energy, log_max_energy, 31)

# Transform the bin edges back to linear scale
linear_bin_edges = 10**log_bin_edges

# Define the bin edges equally
bin_edges_azimuth = np.linspace(np.min(azimuth_P20), np.max(azimuth_P20), 31)
bin_edges_zenith = np.linspace(np.min(zenith_P20), np.max(zenith_P20), 91)

# Filter out the azimuth and zenith values that have been detected so have flag ==1
detected_azimuth_P = [azimuth for azimuth, flag in zip(azimuth_P20, flags_P20) if flag == 1]
detected_zenith_P = [zenith for zenith, flag in zip(zenith_P20, flags_P20) if flag == 1]
detected_zenith_P10 = [zenith for zenith, flag in zip(zenith_P10, flags_P10) if flag == 1]
detected_zenith_P4 = [zenith for zenith, flag in zip(zenith_P4, flags_P4) if flag == 1]
detected_zenith_T = [zenith for zenith, flag in zip(zenith_T20, flags_T20) if flag == 1]
detected_zenith_T10 = [zenith for zenith, flag in zip(zenith_T10, flags_T10) if flag == 1]
detected_zenith_T4 = [zenith for zenith, flag in zip(zenith_T4, flags_T4) if flag == 1]
detected_zenith_L = [zenith for zenith, flag in zip(zenith_L20, flags_L20) if flag == 1]
detected_zenith_L10 = [zenith for zenith, flag in zip(zenith_L10, flags_L10) if flag == 1]
detected_zenith_L4 = [zenith for zenith, flag in zip(zenith_L4, flags_L4) if flag == 1]

# Calculate the distance between the event and the centre of the detector
x_c, y_c, z_c = zip(*location_P20)
z_c = [i + 500 for i in z_c]

distance_P = [np.sqrt((x**2 + y**2 + z**2)) / 1000 for x, y, z in zip(x_c, y_c, z_c)]
detected_distance_P = [distance for distance, flag in zip(distance_P, flags_P20) if flag == 1]

# determine bin edges
bin_edges_distance = np.linspace(np.min(distance_P), np.max(distance_P), 31)

###################################### Now try to calculate the effective volume ###############################################

# Calculate the histograms using numpy.histogram
hist_all_P, _ = np.histogram(energy_P20, bins=linear_bin_edges)
hist_detected_P, _ = np.histogram(detected_energy_P, bins=linear_bin_edges)
hist_all_P10, _ = np.histogram(energy_P10, bins=linear_bin_edges)
hist_detected_P10, _ = np.histogram(detected_energy_P10, bins=linear_bin_edges)
hist_all_P4, _ = np.histogram(energy_P4, bins=linear_bin_edges)
hist_detected_P4, _ = np.histogram(detected_energy_P4, bins=linear_bin_edges)
hist_all_T, _ = np.histogram(energy_T20, bins=linear_bin_edges)
hist_detected_T, _ = np.histogram(detected_energy_T, bins=linear_bin_edges)
hist_all_T10, _ = np.histogram(energy_T10, bins=linear_bin_edges)
hist_detected_T10, _ = np.histogram(detected_energy_T10, bins=linear_bin_edges)
hist_all_T4, _ = np.histogram(energy_T4, bins=linear_bin_edges)
hist_detected_T4, _ = np.histogram(detected_energy_T4, bins=linear_bin_edges)
hist_all_L, _ = np.histogram(energy_L20, bins=linear_bin_edges)
hist_detected_L, _ = np.histogram(detected_energy_L, bins=linear_bin_edges)
hist_all_L10, _ = np.histogram(energy_L10, bins=linear_bin_edges)
hist_detected_L10, _ = np.histogram(detected_energy_L10, bins=linear_bin_edges)
hist_all_L4, _ = np.histogram(energy_L4, bins=linear_bin_edges)
hist_detected_L4, _ = np.histogram(detected_energy_L4, bins=linear_bin_edges)

# Ensure you avoid division by zero
division_result_P = np.divide(hist_detected_P, hist_all_P, where=hist_all_P!=0)
division_result_P10 = np.divide(hist_detected_P10, hist_all_P10, where=hist_all_P10!=0)
division_result_P4 = np.divide(hist_detected_P4, hist_all_P4, where=hist_all_P4!=0)
division_result_T = np.divide(hist_detected_T, hist_all_T, where=hist_all_T!=0)
division_result_T10 = np.divide(hist_detected_T10, hist_all_T10, where=hist_all_T10!=0)
division_result_T4 = np.divide(hist_detected_T4, hist_all_T4, where=hist_all_T4!=0)
division_result_L = np.divide(hist_detected_L, hist_all_L, where=hist_all_L!=0)
division_result_L10 = np.divide(hist_detected_L10, hist_all_L10, where=hist_all_L10!=0)
division_result_L4 = np.divide(hist_detected_L4, hist_all_L4, where=hist_all_L4!=0)

# Calculate the generation volume which is a square
Rgen = 17.5                          # km
Hgen = 1.5                           # km
Vgen = (2 * Rgen)**2 * Hgen
print(f"The volume of the can volume is {round(Vgen)} km3.")

# Calculate the effective volume
Veff_P = division_result_P * Vgen
Veff_P10 = division_result_P10 * Vgen
Veff_P4 = division_result_P4 * Vgen
Veff_T = division_result_T * Vgen
Veff_T10 = division_result_T10 * Vgen
Veff_T4 = division_result_T4 * Vgen
Veff_L = division_result_L * Vgen
Veff_L10 = division_result_L10 * Vgen
Veff_L4 = division_result_L4 * Vgen

# Plot the Veff
plt.figure(figsize=(6, 5))
plt.step(linear_bin_edges[:-1], Veff_P, linestyle='-', color='crimson', label='PanDepth, NMin=20')
plt.step(linear_bin_edges[:-1], Veff_P10, linestyle='-', color='orange', label='PanDepth, NMin=10')
plt.step(linear_bin_edges[:-1], Veff_P4, linestyle='-', color='gold', label='PanDepth, NMin=4')
plt.step(linear_bin_edges[:-1], Veff_T, linestyle='-', color='blue', label='NMin=20')
plt.step(linear_bin_edges[:-1], Veff_T10, linestyle='-', color='dodgerblue', label='NMin=10')
plt.step(linear_bin_edges[:-1], Veff_T4, linestyle='-', color='cyan', label='NMin=4')
#plt.step(linear_bin_edges[:-1], Veff_L, linestyle='-', color='Purple', label='Lahman PanDepth NMin=20')
#plt.step(linear_bin_edges[:-1], Veff_L10, linestyle='-', color='violet', label='Lahman PanDepth NMin=10')
#plt.step(linear_bin_edges[:-1], Veff_L4, linestyle='-', color='pink', label='Lahman PanDepth NMin=4')

plt.xscale('log')
plt.yscale('log')
#plt.xlim(10**(17.8), 10**(22.2))
plt.xlabel('Neutrino Energy [eV]')
plt.ylabel('Effective Volume [$km^3$]')
plt.title('Effective Volume')
plt.grid(True, which='both', linestyle='--', color='gray', alpha=0.3)

plt.legend()
plt.savefig("Veff Tanganyika.jpg")

plt.show()

print(Veff_L4)

###################################### Now try to calculate the Angular Sensitivity ###########################################

# Calculate the histograms using numpy.histogram
zenith_hist_all_P, _ = np.histogram(zenith_P20, bins=bin_edges_zenith)
zenith_hist_detected_P, _ = np.histogram(detected_zenith_P, bins=bin_edges_zenith)
zenith_hist_all_P10, _ = np.histogram(zenith_P10, bins=bin_edges_zenith)
zenith_hist_detected_P10, _ = np.histogram(detected_zenith_P10, bins=bin_edges_zenith)
zenith_hist_all_P4, _ = np.histogram(zenith_P4, bins=bin_edges_zenith)
zenith_hist_detected_P4, _ = np.histogram(detected_zenith_P4, bins=bin_edges_zenith)
zenith_hist_all_T, _ = np.histogram(zenith_T20, bins=bin_edges_zenith)
zenith_hist_detected_T, _ = np.histogram(detected_zenith_T, bins=bin_edges_zenith)
zenith_hist_all_T10, _ = np.histogram(zenith_T10, bins=bin_edges_zenith)
zenith_hist_detected_T10, _ = np.histogram(detected_zenith_T10, bins=bin_edges_zenith)
zenith_hist_all_T4, _ = np.histogram(zenith_T4, bins=bin_edges_zenith)
zenith_hist_detected_T4, _ = np.histogram(detected_zenith_T4, bins=bin_edges_zenith)
zenith_hist_all_L, _ = np.histogram(zenith_L20, bins=bin_edges_zenith)
zenith_hist_detected_L, _ = np.histogram(detected_zenith_L, bins=bin_edges_zenith)
zenith_hist_all_L10, _ = np.histogram(zenith_L10, bins=bin_edges_zenith)
zenith_hist_detected_L10, _ = np.histogram(detected_zenith_L10, bins=bin_edges_zenith)
zenith_hist_all_L4, _ = np.histogram(zenith_L4, bins=bin_edges_zenith)
zenith_hist_detected_L4, _ = np.histogram(detected_zenith_L4, bins=bin_edges_zenith)

# Ensure you avoid division by zero
zenith_division_result_P = np.divide(zenith_hist_detected_P, zenith_hist_all_P, where=zenith_hist_all_P!=0)
zenith_division_result_P10 = np.divide(zenith_hist_detected_P10, zenith_hist_all_P10, where=zenith_hist_all_P10!=0)
zenith_division_result_P4 = np.divide(zenith_hist_detected_P4, zenith_hist_all_P4, where=zenith_hist_all_P4!=0)
zenith_division_result_T = np.divide(zenith_hist_detected_T, zenith_hist_all_T, where=zenith_hist_all_T!=0)
zenith_division_result_T10 = np.divide(zenith_hist_detected_T10, zenith_hist_all_T10, where=zenith_hist_all_T10!=0)
zenith_division_result_T4 = np.divide(zenith_hist_detected_T4, zenith_hist_all_T4, where=zenith_hist_all_T4!=0)
zenith_division_result_L = np.divide(zenith_hist_detected_L, zenith_hist_all_L, where=zenith_hist_all_L!=0)
zenith_division_result_L10 = np.divide(zenith_hist_detected_L10, zenith_hist_all_L10, where=zenith_hist_all_L10!=0)
zenith_division_result_L4 = np.divide(zenith_hist_detected_L4, zenith_hist_all_L4, where=zenith_hist_all_L4!=0)

# Plot the Angular Dependence
plt.figure(figsize=(6, 5))
plt.step(bin_edges_zenith[:-1], zenith_division_result_P, linestyle='-', color='crimson', label='PanDepth, NMin=20')
plt.step(bin_edges_zenith[:-1], zenith_division_result_P10, linestyle='-', color='orange', label='PanDepth, NMin=10')
plt.step(bin_edges_zenith[:-1], zenith_division_result_P4, linestyle='-', color='gold', label='PanDepth, NMin=4')
plt.step(bin_edges_zenith[:-1], zenith_division_result_T, linestyle='-', color='blue', label='NMin=20')
plt.step(bin_edges_zenith[:-1], zenith_division_result_T10, linestyle='-', color='dodgerblue', label='NMin=10')
plt.step(bin_edges_zenith[:-1], zenith_division_result_T4, linestyle='-', color='cyan', label='NMin=4')
#plt.step(bin_edges_zenith[:-1], zenith_division_result_L, linestyle='-', color='purple', label='Lahman PanDepth NMin=20')
#plt.step(bin_edges_zenith[:-1], zenith_division_result_L10, linestyle='-', color='violet', label='Lahman PanDepth NMin=10')
#plt.step(bin_edges_zenith[:-1], zenith_division_result_L4, linestyle='-', color='pink', label='Lahman PanDepth NMin=4')
plt.xlabel('Zenith Angle [degrees]')
plt.ylabel('Detection Efficiency [-]')
plt.title('Zenith Angular Sensitivity')
plt.grid(True, which='both', linestyle='--', color='gray', alpha=0.3)

plt.legend()
plt.savefig("Zenith Tanganyika.jpg")

plt.show()
print(zenith_division_result_P4)

print(sum(zenith_division_result_P))
```

```python
import numpy as np
import matplotlib.pyplot as plt

def pressure_amplitude(r):
    return np.where(r <= 1000, Por / np.sqrt(r), Por / r)

Por = 10000
# Test the function with some sample distances
distances1 = np.linspace(1, 10000, 100)
pressures1 = pressure_amplitude(distances1)

def smooth_transition(r, r_transition, Por_far, Por_near):
    """
    Compute the pressure using a smooth transition between far and near fields.
    
    Parameters:
        r (float or numpy array): Distance from the source.
        r_transition (float): Transition distance from far to near field.
        Por_far (float): Pressure amplitude in the far field.
        Por_near (float): Pressure amplitude in the near field.
        
    Returns:
        numpy array: Pressure values corresponding to the distances.
    """
    # Define the transition function
    def transition_func(r):
        return (1 - np.tanh((r - r_transition) / (r_transition / 4))) / 2
    
    # Apply the transition function
    transition = transition_func(r)
    
    # Compute the pressure using the transition
    pressure_far = Por_far / r  # Pressure in the far field
    pressure_near = Por_near / np.sqrt(r)  # Pressure in the near field
    pressure = pressure_far * (1 - transition) + pressure_near * transition
    
    return pressure

# Define parameters
r_transition = 1000  # Transition distance from far to near field
Por_far = 10000  # Pressure amplitude in the far field
Por_near = 10000  # Pressure amplitude in the near field

# Generate example distances
distances = np.linspace(1, 5000, 1000)

# Compute the pressure values for the example distances
pressures = smooth_transition(distances, r_transition, Por_far, Por_near)

# Plot distance vs. pressure
plt.figure(figsize=(10, 6))
plt.plot(distances, pressures, color='blue')
plt.plot(distances1, pressures1, color='dodgerblue')
plt.axvline(250, color='orange')
plt.axvline(1000, color='orange')
plt.yscale("log")
plt.xlabel('Distance [m]')
plt.ylabel('Pressure [mPa]')
plt.title('Pressure vs. Distance with Smooth Transition')
plt.grid(True)
plt.show()
```
